```python
import matplotlib.pyplot as plt
import sys

def show_img(path):
    data = cv2.imread(path, cv2.IMREAD_COLOR)
    if data is None:
        sys.exit("Data is empty!")
    data = data[:, :, ::-1]
    data = cv2.cvtColor(data, cv2.COLOR_BGR2RGB)
    plt.imshow(data)
    plt.axis('off')
    plt.show()
```

```python
show_img('sign2.png')
```

```python
show_img('sign5.png')
```

```python
show_img('sign3.png')
```

```python
show_img('sign4.png')
```

```python
from numpy.linalg import norm

def cos_sim(x, y):
    x = x.flatten()
    y = y.flatten()

    x = x.astype(np.float64)
    y = y.astype(np.float64)

    cos_sim = np.dot(x, y) / (norm(x)*norm(y))
    return cos_sim
```

```python
cos_sim(img1, img1)
```

```python
cos_sim(img1, img3)
```

```python
cos_sim(img1, img4)
```
