```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

# read a grayscale image and plot it
data1 = cv2.imread('flowers.jpg', 1)
print(type(data1[0,0,0]))

data1 = cv2.cvtColor(data1, cv2.COLOR_BGR2RGB)
height, width, channel = data1.shape

data2 = np.zeros( (height, width, channel), np.uint8)
print(data2.shape)

for i in range(height):
    for j in range(width):
        for c in range(channel):
            # for each pixel
            data2[i,j,c] = data1[i,j,c]

print(data2[0,0,0])
print(type(data2[0,0,0]))
plt.imshow(data2)
plt.show()
```

##**Create black and noise images**

```python
# create a black image

import cv2
import numpy as np

height = 800
width = 500
image = [[0]*width for _ in range(height)]

cv2.imwrite('2.black_image.jpg', np.array(image))
```

```python
# create a noise image

import cv2
import numpy as np
import random

height = 800
width = 500
image = [[0]*width for _ in range(height)]

# set noise to each pixel; a random integer in [0, 2550]
for i in range(height):
    for j in range(width):
        image[i][j] = random.randint(0, 255)

cv2.imwrite('2.noise_image.jpg', np.array(image))
```

#**Simple Applications**

##**Copy an image**

```python
# copy image using list
import cv2
import numpy as np

# read an image
image = cv2.imread('nature.jpg', 0)
height, width = image.shape

# create a 2D list
copy_image = [[0]*width
              for _ in range(height)]

# copy
for i in range(height):
    for j in range(width):
        copy_image[i][j] = image[i][j]

# save an image from a list
cv2.imwrite('copy.jpg', np.array(copy_image))
```

```python
# copy image using array
import cv2
import numpy as np

# read an image
image = cv2.imread('nature.jpg', 0)
height, width = image.shape

# create a 2D array
array_shape = (height, width)
copy_image = np.zeros(array_shape)

# copy
for i in range(height):
    for j in range(width):
        copy_image[i][j] = image[i][j]

# save an image from a list
cv2.imwrite('copy.jpg',
            np.array(copy_image))
```

**For Color Image**

```python
# copy image using numpy
import cv2
import numpy as np

# read an image
image = cv2.imread('flowers.jpg', 1)
height, width, channel = image.shape

# create an array
array_shape = (height, width, channel)
copy_image = np.zeros(array_shape)

# copy
for i in range(height):
    for j in range(width):
        for c in range(channel):
            copy_image[i][j][c] = image[i][j][c]

# save an image from a list
cv2.imwrite('copy.jpg',
            np.array(copy_image))
```

##**Change illuminations**

```python
# Copy and increase image brightness (1)

import cv2
import numpy as np

# read an image
image = cv2.imread('nature.jpg', 0)
height, width = image.shape

# create a 2D list
copy_image = [[0]*width for _ in range(height)]

# copy and increase image brightness
value = 50
for i in range(height):
    for j in range(width):
        copy_image[i][j] = min(255, image[i][j] + value)

# save an image from a list
cv2.imwrite('3.nature_increase_50.jpg', np.array(copy_image))
```

```python
import numpy as np
import cv2

bg = cv2.imread('background.png', 1)
bg = cv2.resize(bg, (640, 480))

img = cv2.imread('StillImage.png', 1)
img = cv2.resize(img, (640, 480))

difference = cv2.absdiff(bg, img)
_, difference_binary = cv2.threshold(
    difference, 20, 255, cv2.THRESH_BINARY
)

new_bg = cv2.imread('FakeBackground.png')
new_bg = cv2.resize(new_bg, (640, 480))

output = np.where(difference_binary==0, new_bg, img)
cv2.imwrite('output.png', output)
```

```python
import numpy as np
import cv2

bg = cv2.imread('background.png', 1)
bg = cv2.resize(bg, (640, 480))

img = cv2.imread('StillImage.png', 1)
img = cv2.resize(img, (640, 480))

new_bg = cv2.imread('FakeBackground.png')
new_bg = cv2.resize(new_bg, (640, 480))

difference = np.abs(img-bg)
difference_binary = np.where(difference<20, 0, 255)

output = np.where(difference_binary==0, new_bg, img)
cv2.imwrite('output.png', output)
```
