## Bag of Words (NLP)

### Bài tập:
 Tạo Bag-Of-Word cho tập dataset sau: corpus = ["Tôi thích môn Toán",
"Tôi thích AI", "Tôi thích âm nhạc"]. Sau đó tạo list có tên vector để lưu vector sau khi thực hiện bước Tokenization đoạn văn bản sau: Tôi thích AI thích Toán, biết Bag-Of-Word được sắp theo thứ tự tăng dần

```python
def get_vocabulary(corpus:list[str]):
  tokenize_cached = []
  vocabulary = []
  bow_list_of_corpus = []
  for i in range(len(corpus)):
    words_list = corpus[i].split(" ")
    words_list.sort()
    tokenize_cached.extend([words_list])
    for i in range(len(words_list)):
      if words_list[i] not in vocabulary:
        vocabulary.append(words_list[i])
  vocabulary.sort()
  return tokenize_cached, vocabulary
```

```python
def find_word_in_vocabulary(word, vocabulary):
  for i in range(len(vocabulary)):
    if word == vocabulary[i]:
      return i
  return -1
```

```python
def get_bag_of_words(document, vocabulary):
  bow = [0] * len(vocabulary)
  words_list = document.split(" ")
  for i in range(len(words_list)):
    index = find_word_in_vocabulary(words_list[i], vocabulary)
    if index != -1:
      bow[index] += 1
  return bow
```

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"] #list str
tokenize_cached, vocabulary = get_vocabulary(corpus)
print(f"tokenize_cached = {tokenize_cached}")
print(f"vocabulary = {vocabulary}")

doc = "Tôi thích AI thích Toán"
bow = get_bag_of_words(doc, vocabulary)
print(f"Bag of words of sentence {doc} is {bow}")
```

## Interpolation for List (Time-series)

### Bài tập:
Tạo List có tên là lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]. Sau đó,
áp dụng phương pháp nội suy Nearest Neighbor để gắn giá trị None có trong lst_data


```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
if lst_data[0] == None:
  lst_data[0] = lst_data[1]
elif lst_data[-1] == None:
  lst_data[-1] = lst_data[-2]

for i in range(2, len(lst_data) - 1):
  if lst_data[i] == None:
    if lst_data[i - 1] > lst_data[i + 1]:
      lst_data[i] = lst_data[i + 1]
    else:
      lst_data[i] = lst_data[i - 1]
print(lst_data)
```

## Matrix representation using List

### Bài tập:

```python
def sum(A, B):
  if len(A) != len(B) or len(A[0]) != len(B[0]):
    print("Cannot sum 2 matrixes")
    return -1
  result = []
  # A, B have the same size
  for i in range(len(A)): # row
    row = []
    for j in range(len(A[0])):
      row.append(A[i][j] + B[i][j])
    result.append(row)
  return result
```

```python
def subtract(A, B):
  if len(A) != len(B) or len(A[0]) != len(B[0]):
    print("Cannot sum 2 matrixes")
    return -1
  result = []
  # A, B have the same size
  for i in range(len(A)): # row
    row = []
    for j in range(len(A[0])):
      row.append(A[i][j] - B[i][j])
    result.append(row)
  return result
```

```python
A = [[1,2,3],[4,5,6],[7,8,9]]
B = [[2,4,6],[1,3,5],[1,0,1]]
# 1)
# sum
print(f"A + B = {sum(A, B)}")

# subtract
print(f"A - B = {subtract(A, B)}")
```

```python
def matrix_multiplication(A, B): # A(m,n), B(n,p)
  try:# matrix input like [1,2,3]
    len(A[0])
  except:
    A = [A] # convert input to [[1,2,3]]

  try:
    len(B[0])
  except:
    B = [B]

  if len(B) != len(A[0]):
    print("Please re-check 2 array: Number colums of A must be equal number row of B!")
    return -1
  result = []
  for i in range(len(A)): # row A
    row_result = []
    for m in  range(len(B[0])): # column B
      sum = 0
      for j in range(len(B)): # row B and column A
        sum += A[i][j] * B[j][m]
      row_result.append(sum)
    result.append(row_result)

  return result
```

 #### Chuẩn hóa dữ liệu (Normalization)

```python
import math

def find_min_max(lst_data):
  min = max = lst_data[0]
  for i in range(1, len(lst_data)):
    if min > lst_data[i]:
      min = lst_data[i]
    if max < lst_data[i]:
      max = lst_data[i]
  return min, max

def normalize(lst_data):
  min, max = find_min_max(lst_data)
  return [(e - min) / (max - min) for e in lst_data]

test_cases = [
    [11, 18, 24, 30, 36],#test
    [50, 100, 150, 200, 250],
    [3, 5, 7, 9, 11]
]

for i, test in enumerate(test_cases):
  result = normalize(test)
  print(f"Test {i+1}: {result}")
```

 #### Trung bình động (Moving Average)

```python
# def MA_at_t(lst, t, window_size):
#   MAt = 0
#   for i in range(t, t + window_size):
#     MAt += lst[i]
#   MAt /= window_size
#   return MAt
```

```python
def moving_average(lst, k):
  MA = []
  for t in range(len(lst) - k + 1):
    MAt = 0
    for i in range(t, t + k):
      MAt += lst[i]
    MAt /= k
    MA.append(MAt)
  return MA

test_cases = [
    ([1, 2, 3, 4, 5, 6, 7, 8, 9], 3),
    ([10, 20, 30, 40, 50, 60, 70], 4),
    ([5, 10, 15, 20, 25], 2)
]

for i, (test, k) in enumerate(test_cases):
  result = moving_average(test, k)
  print(f"Test {i+1}: {result}")
```

 #### Tích vô hướng của hai vector

```python
def dot_product(v1, v2):
  result = 0
  for i in range(len(v1)):
    result += v1[i] * v2[i]
  return result

test_cases = [
 ([1, 2, 3], [4, 5, 6]),
 ([2, 4, 6], [1, 3, 5]),
 ([0, 1, 2], [3, 4, 5])
]

for i, (v1, v2) in enumerate(test_cases):
  result = dot_product(v1, v2)
  print(f"Test {i+1}: {v1} dot product {v2} = {result}")
```

 #### Mô phỏng Perceptron

```python
def relu(x):
  if x > 0.0:
    return x
  else:
    return 0.0
```

```python
def perceptron_relu(weights, inputs, bias):
  y = 0.0
  for i in range(len(weights)):
    y += weights[i] * inputs[i]
  y += bias
  return relu(y)

test_cases = [
    ([0.5, -0.6, 0.8], [1.0, 0.0, 1.0], -0.3),
    ([0.2, 0.5, -0.4], [1.0, 2.0, -1.0], 0.1),
    ([1.0, -1.0, 0.5], [0.5, 1.0, -0.5], -0.2)
]

for i, (weights, inputs, bias) in enumerate(test_cases):
  result = perceptron_relu(weights, inputs, bias)
  print(f"Test {i+1}: {result}")
```

## Computing mean for a list of numbers

# Day 10: Practice with List - Positional Embedding

## Bài tập
Giả sử chúng ta có một câu với độ dài tối đa là 10 từ (max_len = 10) và vector embedding có độ dài là 16 (dmodel = 16). Hãy viết hàm tính Positional embedding cho câu này.


Version không dùng Numphy

```python
import math

def position_embedding(pos, i, embed_size):
  if i % 2 == 0:
    return math.sin(pos / math.pow(1000,  (2 * i) / embed_size))
  else:
    return math.cos(pos / math.pow(1000,  (2 * i) / embed_size))


def create_position_matrix(seq_length, embed_size):
  position_matrix = [[0 for _ in range(embed_size)] for _ in range(seq_length)]
  # pos of word = 0 => 9
  # i: length of vector embedding = 0 => 15
  for pos in range(seq_length):
    for i in range(embed_size):
      position_matrix[pos][i] = position_embedding(pos, i, embed_size=embed_size)

  return position_matrix

#Test
seq_length = 10 #max_len words
embed_size = 16 #dmodel
position_matrix = create_position_matrix(seq_length, embed_size)
print("Display as list matrix: ")
print(position_matrix)
```
