## **12. Filtering bằng regex**

```python
import re
```

```python
data.loc[data["Name"].str.contains("Mega")].head(3)
```

```python
data.loc[data["Type 1"].str.contains("Fire|grass", regex=True)].head(6)
```

```python
data.loc[data["Type 1"].str.contains("fire|grass", flags=re.I, regex=True)].head(6)
```

```python
data.loc[data["Name"].str.contains("pi[a-z]*", flags=re.I, regex=True)].head(3)
```

```python
data.loc[data["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)].head(3)
```

```python
data.loc[data["Name"].str.contains("^pi.*", flags=re.I, regex=True)].head(3)
```

## **14. Groupby**

```python
data.groupby("Type 1")[data.select_dtypes(include="number").columns].mean().sort_values(
    "Defense", ascending=False
).head(10)
```

```python
data.count()
```

```python
data.groupby(["Type 1", "Type 2"]).count()
```

```python
data.groupby(["Type 1"]).sum()
```

```python
df = data.groupby("Type 1")[["Attack", "Defense"]].apply(lambda x: x.sum() / len(x))
df.head(3)
```

```python
df = data.groupby(["Type 1"]).agg(
    {"Attack": "mean", "Defense": "sum", "Speed": lambda x: len(x)}
)
df.head(3)
```
