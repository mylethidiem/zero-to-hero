```python
import seaborn as sns
import cv2
import random

sns.set_theme()

# load img
# img = cv2.imread('image.jpeg', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('image.jpeg', 0)
height, width = img.shape

# ...
num_of_samples = 10000
sample_size = 50

means = []
for _ in range(num_of_samples):
    mean = 0
    for _ in range(sample_size):
        # generate a random point
        x = random.randint(0, height - 1)
        y = random.randint(0, width - 1)

        # get pixel value and add to mean
        mean = mean + img[x][y]
    # normalize to get mean
    mean = mean / sample_size

    # add to list
    means.append(mean)
```

```python
len(means)
print(means)
```

```python
sns.displot(means, kde=True)
```

```python
import seaborn as sns
import cv2
import numpy as np # Import numpy
import matplotlib.pyplot as plt
sns.set_theme()

# load image
img = cv2.imread('image.jpeg', cv2.IMREAD_GRAYSCALE)
height, width = img.shape
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
print(img)
```

```python
# ...
num_of_samples = 10000
sample_size = 50

# --- Corrected Code ---

# 1. Generate all random coordinates at once using NumPy's random module
# This is much faster than calling random.randint() in a loop.
# Generate random value from (0, height/width)
# with output array shape size=(num_of_samples, sample_size)
random_x = np.random.randint(0, height, size=(num_of_samples, sample_size))
random_y = np.random.randint(0, width, size=(num_of_samples, sample_size))

# 2. Use advanced indexing to get all pixel values in one shot
# This creates a new array 'pixels' with the same shape as our random coordinates.
pixels = img[random_x, random_y]

# 3. Calculate the mean of each sample (each row) in one vectorized operation
# axis=1 tells mean() to calculate the average across the columns for each row.
means = np.mean(pixels, axis=1)

# --- End Corrected Code ---

print(pixels)
print(pixels.shape)
```

```python
print(len(means))
print(means)
print(means.shape)
```

```python
sns.displot(means, kde=True)
```
