## Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x_data = np.arange(0, 7, 0.1)
```

```python
def gaussian_function(data, mean, var):
    return (1.0 / np.sqrt(2 * math.pi * var)) * np.exp(-(data - mean) ** 2 / (2 * var))

mean1 = 1.5666666666666667
var1  = 0.3288888888888889
y_data1 = gaussian_function(x_data, mean1, var1)
print(y_data1)
```

```python
mean2 = 3.733333333333333
var2  = 0.2722222222222222
y_data2 = gaussian_function(x_data, mean2, var2)
print(y_data2)
```

```python
plt.plot(x_data, y_data1)
plt.plot(x_data, y_data2)
```

## Gaussian Naive Bayes 1D

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

import pandas as pd

# Read data iris
iris = pd.read_csv('iris_1D.csv')

iris.head()
```

```python
# Seperating the data into dependent and independent variables
X_train = iris.iloc[:, :-1].values
y_train = iris.iloc[:, -1].values
print("X_train[:, 0]: ", X_train[:, 0])
print("y_train: ", y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np

# Train model
classifier = GaussianNB()
clf = classifier.fit(X_train, y_train)

# predict
x_test = np.array([[3.0]])
val = clf.predict_proba(x_test)
print("\nval ", val)
print("\nclass_count_ ", clf.class_count_)
print("\nclass_prior_ ", clf.class_prior_)
print("\nclasses_ ", clf.classes_)
print("\nvar_ ", clf.var_)
print("\ntheta_ ", clf.theta_)
```

### Compute manually

```python
import math
import numpy as np

def gaussian_function(data, mean, var):
    return (1.0 / np.sqrt(2 * math.pi * var)) * np.exp(-(data - mean) ** 2 / (2 * var))

def show_result(mean=0, var=0, prob=np.array([])):
    print("mean = ", mean) if mean else print()
    print("var = ", var) if var else print()
    print("prob = ", prob) if prob.size != 0 else print()
```

```python
# Compute Gaussian for Group 1
data_c0 = np.array([1.4, 1.0, 1.3, 1.9, 2.0])
mean_c0 = data_c0.mean()
var_c0 = data_c0.var()
show_result(mean_c0, var_c0)
```

```python
# compute Gaussian for group 2
data_c1 = np.array([3.8, 4.1, 3.9, 4.2, 3.4])
mean_c1 = data_c1.mean()
var_c1  = data_c1.var()
prob_c1 = gaussian_function(data_c1, mean_c1, var_c1)
show_result(mean_c1, var_c1, prob_c1)
```

```python
# predict manually
x_test = np.array([[3.0]])
pdf_c0_x = gaussian_function(x_test, mean_c0, var_c0)
pdf_c1_x = gaussian_function(x_test, mean_c1, var_c1)

# normalize
total = pdf_c0_x*0.5 + pdf_c1_x*0.5

#explain
pdf_c0_x = pdf_c0_x*0.5/total
pdf_c1_x = pdf_c1_x*0.5/total
print(pdf_c0_x)
print(pdf_c1_x)
```
