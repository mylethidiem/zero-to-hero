```python
import numpy as np
import cv2

img = cv2.imread('scene.png', 0)

equ = cv2.equalizeHist(img)
res = np.hstack((img,equ))

cv2.imwrite('scene_equ.jpg', equ)
#cv2.imwrite('img_result3.jpg', res)
```

```python
img = cv2.imread('Book4.png', 1)
print(img.shape)

img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

# equalize the histogram of the Y channel
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])

# convert the YUV image back to RGB format
img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

cv2.imwrite('img_result1_4.jpg', img_output)
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('Book4.png')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Book1_equ.jpg',0)

hist,bins = np.histogram(img.flatten(),256,[0,256])

plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
plt.hist(img.flatten(),256,[0,256], color = 'r', density=True)
plt.xlim([0,256])
plt.show()
```

```python
cdf = hist.cumsum()
cdf_normalized = cdf/cdf.max()

plt.plot(cdf_normalized, color = 'g')
plt.xlim([-1,256])
plt.show()
```
