```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image2.png',flags=cv2.IMREAD_GRAYSCALE)
# cv2.IMREAD_GRAYSCALE: 0
# cv2.IMREAD_UNCHANGED: (or -1)
# cv2.IMREAD_COLOR: (or 1), default
# display the image
plt.figure(figsize=(4,2))
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image1.png')

# convert the image from BGR to RGB
# image_rgb = image[:, :, ::-1]
# image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb = image[:, :, [2, 1, 0]]
# Explain:
# :, : => all elements along the first and the second axis (height, width)
# third axis is the color channel
# Index 0: Blue channel
# Index 1: Green channel
# Index 2: Red channel

plt.imshow(image_rgb)
```
