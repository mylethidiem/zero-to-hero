### Temperature

```python
import pandas as pd
import numpy as np

df = pd.read_csv('temperature-1d.csv')
data = df.to_numpy()
df.head()

```

```python
temp = data[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(temp)
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.show()
```

```python
temp_rs = temp.reshape(-1, 24)#colapse result to 24 cols
#representing the 24 hours in a day
# show from the result we assume that we have 10 days with 24 hours
print(temp_rs.shape)
# temp_rs
```

```python
print(temp.shape)
# temperory calculate temperature for each day
# axis=1: Tells the function to calculate
# the average horizontally across the columns for each row.
mean_24 = np.mean(temp_rs, axis=1)
# The result, mean_24, is a 1D array
# where each element is the average temperature for a single day.
print(mean_24[:10])
print(mean_24.shape)

# takes each daily average in mean_24 and repeats it 24 times.
result = np.repeat(mean_24, 24)
print(result.shape)
# Final result is a 1D array with the same lenght as the original temp array
# contains the daily average temperature repeated for every hour of that day

```

```python
import matplotlib.pyplot as plt
#%matplotlib qt

plt.plot(temp, label='Original Temperature')
plt.plot(result, label='Mean Temperature')
plt.xlabel('Index')
plt.ylabel('Temperature')
plt.title('Temperature Data and Mean Temperature')
plt.legend()
plt.show()
```

### Encoding

```python
import numpy as np

# num labels
K = 3

# num sampels
N = 9

labels = [0, 0, 0, 1, 1, 1, 2, 2, 2]
labels = np.array(labels)
print(labels)

# create an NxK array
labels_onehot = np.zeros((N, K))

# one-hot encoding
labels_onehot[np.arange(N), labels] = 1
print(labels_onehot)
```
