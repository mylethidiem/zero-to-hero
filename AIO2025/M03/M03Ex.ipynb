```python
# Convert centers to vectors
c1_vector = [0] * len(dictionary_)
for word in c1.lower().split():
    if word in dictionary_:
        index = dictionary_.index(word)
        c1_vector[index] = 1

c2_vector = [0] * len(dictionary_)
for word in c2.lower().split():
    if word in dictionary_:
        index = dictionary_.index(word)
        c2_vector[index] = 1

# Calculate distances to both centers for each point and assign clusters
cluster_vector = []
for point in X:
    dist_to_c1 = np.sqrt(np.sum((np.array(c1_vector) - point)**2))
    dist_to_c2 = np.sqrt(np.sum((np.array(c2_vector) - point)**2))
    # Assign to cluster 0 if closer to c1, cluster 1 if closer to c2
    cluster = 0 if dist_to_c1 < dist_to_c2 else 1
    cluster_vector.append(cluster)

print("Cluster assignments:")
for doc, cluster in zip(corpus, cluster_vector):
    print(f"{doc}: Cluster {cluster}")

print("\nCluster vector:")
print(np.array(cluster_vector))
```

```python
import numpy as np
from collections import Counter
import gradio as gr

# Tính Entropy
def entropy(y):
    counter = Counter(y)
    probs = [count/len(y) for count in counter.values()]
    return -sum(p * np.log2(p) for p in probs)

# Tính Information Gain
def information_gain(X, y, feature_idx):
    parent_entropy = entropy(y)

    # Tính entropy cho mỗi giá trị của feature
    values = set(X[:, feature_idx])
    weighted_entropy = 0
    for value in values:
        mask = X[:, feature_idx] == value
        child_y = y[mask]
        weight = len(child_y) / len(y)
        weighted_entropy += weight * entropy(child_y)

    return parent_entropy - weighted_entropy

class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature  # index của feature để split
        self.threshold = threshold  # giá trị của feature để split
        self.left = left  # node con trái
        self.right = right  # node con phải
        self.value = value  # giá trị dự đoán tại node lá

class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.root = None
        self.feature_names = None

    def fit(self, X, y, feature_names=None):
        self.feature_names = feature_names
        self.n_classes = len(np.unique(y))
        self.root = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape

        # Điều kiện dừng
        if (self.max_depth is not None and depth >= self.max_depth) or len(np.unique(y)) == 1:
            return Node(value=Counter(y).most_common(1)[0][0])

        # Tìm feature tốt nhất để split
        best_feature = None
        best_gain = -1

        for feature_idx in range(n_features):
            gain = information_gain(X, y, feature_idx)
            if gain > best_gain:
                best_gain = gain
                best_feature = feature_idx

        if best_gain == 0:
            return Node(value=Counter(y).most_common(1)[0][0])

        # Tạo node và chia dữ liệu
        node = Node(feature=best_feature)

        unique_values = np.unique(X[:, best_feature])
        for value in unique_values:
            mask = X[:, best_feature] == value
            if sum(mask) > 0:
                child_X = X[mask]
                child_y = y[mask]
                if node.threshold is None:
                    node.threshold = value
                    node.left = self._grow_tree(child_X, child_y, depth + 1)
                else:
                    node.right = self._grow_tree(child_X, child_y, depth + 1)

        return node

    def predict(self, X):
        return np.array([self._traverse_tree(x, self.root) for x in X])

    def _traverse_tree(self, x, node):
        if node.value is not None:
            return node.value

        if x[node.feature] == node.threshold:
            return self._traverse_tree(x, node.left)
        return self._traverse_tree(x, node.right)

    def visualize(self, node=None, depth=0, prefix=""):
        if node is None:
            node = self.root

        result = ""

        if node.value is not None:
            result += prefix + f"Predict: {node.value}\n"
            return result

        feature_name = f"Feature {node.feature}" if self.feature_names is None else self.feature_names[node.feature]
        result += prefix + f"{feature_name} = {node.threshold}?\n"

        result += self.visualize(node.left, depth + 1, prefix + "├── Yes: ")
        result += self.visualize(node.right, depth + 1, prefix + "└── No: ")

        return result

# Tạo giao diện Gradio
def train_and_visualize(X_str, feature_names_str):
    try:
        # Chuyển đổi input string thành numpy array
        X_lines = X_str.strip().split('\n')
        X = np.array([line.strip().split(',') for line in X_lines])

        # Tách features và labels
        features = X[:, :-1]
        labels = X[:, -1]

        # Chuyển đổi tên features
        feature_names = feature_names_str.strip().split(',')

        # Huấn luyện cây quyết định
        tree = DecisionTree(max_depth=5)
        tree.fit(features, labels, feature_names)

        # Trả về visualization
        return tree.visualize()
    except Exception as e:
        return f"Error: {str(e)}"

# Tạo giao diện
iface = gr.Interface(
    fn=train_and_visualize,
    inputs=[
        gr.Textbox(label="Enter training data (CSV format, last column is label)"),
        gr.Textbox(label="Enter feature names (comma-separated)")
    ],
    outputs=gr.Textbox(label="Decision Tree Visualization"),
    title="Decision Tree Visualizer",
    description="Enter your training data and feature names to visualize the decision tree"
)

# Chạy giao diện
# Add share=True to create a public link
iface.launch(share=True)
```
