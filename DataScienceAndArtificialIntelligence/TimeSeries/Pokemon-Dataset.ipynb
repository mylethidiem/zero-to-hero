## **14. Groupby**

```python
data.groupby('Type 1')[data.select_dtypes(include='number').columns] \
    .mean() \
    .sort_values('Defense', ascending=False) \
    .head(10)
```

```python
data.count()
```

```python
data.groupby(['Type 1', 'Type 2']).count()
```

```python
data.groupby(['Type 1']).sum()
```

```python
df = data.groupby('Type 1')[['Attack', 'Defense']].apply(
    lambda x: x.sum() / len(x)
)
df.head(3)
```

```python
df = data.groupby(['Type 1']).agg({'Attack': 'mean',
                                   'Defense': 'sum',
                                   'Speed': lambda x: len(x)})
df.head(3)
```
