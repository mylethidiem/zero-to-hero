## Bài tập: Khởi tạo List và thao tác thêm xóa sửa trên List
- Câu 1: Tạo mới một List có tên là lst_data, gồm các số chẵn từ 1 đến 12.
- Câu 2: Xóa tất cả các số chia hết cho 3 trong lst_data vừa tạo
- Câu 3: Thêm vào cuối lst_data các số từ 1 đến 3, và thêm vào vị trí index = 3 một chuỗi các số từ 6 đến 8
- Câu 4: Nếu các số trong list lst_data chia hết cho 2 hoặc chia hết cho 5 thì cập nhật thành số 0

```python
lst_data = list(range(2,12 + 1,2))
print("1. List even numbers created: ", lst_data)

lst_data = [lst_data[i] for i in range(len(lst_data)) if lst_data[i] % 3 != 0]
print("\n 2. Erase all numbers divisible by 3 : ", lst_data)

lst_data.extend(list(range(1,4)))
print("\n 3. Add numbers 1 to 3 to the end of lst_data: ", lst_data)

lst_data = lst_data[:3] + list(range(6,9)) + lst_data[3:]
print("and add to the index = 3 position a series of numbers from 6 to 8, final result: ", lst_data)

lst_data = [0 if (x % 2 == 0 or x % 5 == 0) else x for x in lst_data]
print("\n 4. List numbers divisible by 2 or divisible by 5: ", lst_data)
```
