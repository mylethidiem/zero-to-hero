# Day 37: Hệ Thống Kiểm Tra Đạo Văn Bằng TF-IDF và MySQL

### Term Frequency-inverse document frequency
- TF-IDF (term frequency-inverse document frequency) là một kỹ thuật phổ biển trong lĩnh vực xử lý ngôn ngữ tự nhiên và khai khác văn bản. Mục đích của TF-IDF là để đánh giá tầm quan trọng của một từ trong một tài liệu so với toàn bộ tập hợp các tài liệu (corpus)

- Term Frequency (TF) \
Đo lường tần xuất hiện của một từ trong một tài liệu. Công thức tính: \
$$
TF(t,d)=\frac{\text{số lần xuất hiện của từ $t$ trong tài liệu $d$}}{\text{tổng số từ trong tài liệu $d$}}
$$

- Inverse Document Frequency (IDF) \
Đo lường mức độ phổ biến của một từ trong tập hợp các tài liệu. Công thức tính: \
$$
IDF(t,D)=log(\frac{\text{tổng số tài liệu $\left| D \right|$}}{1+\text{số tài liệu chứa từ $t$}})
$$

Trong đó, $|D|$ là tổng số tài liệu và số tài liệu chứa từ $t$ được cộng thêm 1 để tránh chia 0

- TF-IDF \
Là tích của TF và IDF, giúp xác định tầm quan trọng của một từ trong một tài liệu cụ thể, đồng thời giảm thiểu ảnh hưởng của các từ phổ biến nhưng ít mang ý nghĩa. \
$$
TF-IDF(t, d, D) = TF(t, d) \times  IDF(t, D)
$$

TF-IDF giúp phân biệt các từ quan trọng đối với nội dung của một tài liệu so với các từ xuất hiện thường xuyên trong nhiều tài liệu khác nhau nhưng ít mang giá trị thông tin.

```python
import pandas as pd
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Read data
df = pd.read_csv('data.csv')

#Step 2: Preprocess document
def clean_text(text):
  text = re.sub(r'^Subject:\s*','',text, flags=re.IGNORECASE)
  text = text.replace('\r',' ').replace('\n',' ')
  text = re.sub (r'[^a-zA -ZÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚĂĐĨŨƠàáâãèéêìíòóôõùúăđĩũơƯĂẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼỀỀỂẾưăạảấầẩẫậắằẳẵặẹẻẽềềểếỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪễệỉịọỏốồổỗộớờởỡợụủứừỬỮỰỲỴÝỶỸửữựỳỵýỷỹ\s]', '', text)
  text = text.lower()
  text = re.sub(r'\s+',' ', text).strip()
  return text

df['cleaned'] = df['text'].astype(str).apply(clean_text)
text = ' '.join(df['cleaned'])

# Step 3: Create and display WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordClound from preprocessing data', fontsize=16)
plt.show()
```

# Day 39: Phân tích IQR và Phát hiện Outlier trong Dữ liệu Thống kê

Trong phân tích thống kê, IQR (Interquartile Range) hay khoảng tứ phân vị, là một thước đo quan trọng dùng để đánh giá mức độ phân tán của dữ liệu. IQR thể hiện phạm vi chứa 50% số liệu trung tâm của một tập dữ liệu, được xác định bởi hiệu số giữa phần tư thứ ba (Q3) và phần tư thứ nhất (Q1): \
$$IQR = Q3 - Q1$$
IQR có ưu điểm là không bị ảnh hưởng bởi các giá trị cực đoan, nên rất phù hợp để mô tả sự phân bố trung tâm của các tập dữ liệu không phân phối chuẩn.

Cách tính Quartile
Quartile (Tứ phân vị) là các giá trị chia tập dữ liệu đã được sắp xếp thành bốn phần bằng nhau, mỗi phần tương ứng với 25% số quan sát.
1. Sắp xếp dữ liệu: Bước đầu tiên là sắp xếp tập dữ liệu của bạn theo thứ tự tăng dần.
2. Xác định trung vị (Q2):
	2.1 Nếu số lượng phần tử (n) của tập dữ liệu là lẻ, thì trung vị là phần tử ở vị trí (n + 1) / 2.
	2.2 Nếu n là chẵn, thì trung vị là trung bình của hai phần tử ở vị trí n / 2 và (n / 2) + 1.
3. Tính Q1: Q1 là trung vị của nửa dưới của dữ liệu (không bao gồm trung vị nếu n là lẻ).
4. Tính Q3: Q3 là trung vị của nửa trên của dữ liệu (không bao gồm trung vị nếu n là lẻ).

Cho tập dữ liệu: 3, 7, 8, 5, 12, 14, 21, 13, 18

1. Sắp xếp: 3, 5, 7, 8, 12, 13, 14, 18, 21
2. Q2 (Trung vị): Vì n = 9 (lẻ), Q2 là phần tử ở vị trí (9 + 1) / 2 = 5. Vậy Q2 = 12.
3. Q1: Nửa dưới của dữ liệu: 3, 5, 7, 8. Vì có 4 phần tử (chẵn), Q1 là trung bình của hai số ở vị trí 4/2 = 2 và (4/2) + 1 = 3. Vậy Q1 = (5 + 7) / 2 = 6.
4. Q3: Nửa trên của dữ liệu: 13, 14, 18, 21. Tương tự, Q3 = (14 + 18) / 2 = 16.

*Outlier và Phát hiện Outlier bằng IQR*

Outlier (Giá trị ngoại lai) là các điểm dữ liệu có giá trị khác biệt rõ rệt so với xu hướng chung của tập dữ liệu. Chúng có thể do lỗi nhập liệu hoặc là các hiện tượng đặc biệt hiếm gặp. \
Dựa vào IQR, các ngưỡng để xác định outlier được tính như sau:
- Ngưỡng dưới (Lower Bound) = Q1 - 1.5 x IQR
- Ngưỡng trên (Upper Bound) = Q3 + 1.5 x IQR

Nếu một giá trị nằm ngoài khoảng [Lower Bound, Upper Bound] thì được coi là outlier.

? Tại sao sử dụng hệ số 1.5? \
Hệ số 1.5 trong phương pháp IQR được đề xuất bởi John Tukey, người phát triển biểu đồ hộp (boxplot). Lý do chọn hệ số này: \
• Tương đương khoảng 2.7 độ lệch chuẩn trong phân phối chuẩn. \
• Giúp cân bằng giữa việc phát hiện điểm bất thường thật sự và tránh loại bỏ các điểm hợp lệ. \
• Hệ số nhỏ hơn (như 1.0) sẽ loại bỏ quá nhiều giá trị hợp lệ; hệ số lớn hơn (như 3.0) có thể bỏ sót outlier thật sự.

Phân tích IQR và phát hiện outlier là bước quan trọng trong tiền xử lý dữ liệu. Sử dụng ngưỡng 1.5 × IQR là một phương pháp đơn giản nhưng hiệu quả để xác định các giá trị bất thường mà không bị ảnh hưởng bởi dữ liệu cực đoan. Tuy nhiên, việc loại bỏ outlier cần được cân nhắc trong từng trường hợp cụ thể nhằm đảm bảo tính toàn vẹn và ý nghĩa của dữ liệu.

# Day 42: Work with Text Data


Thư viện Gensim là một thư viện Python mạnh mẽ, được thiết kế để xử lý Natural Language Processing (NLP), đặc biệt là trong các bài toán liên quan đến mô hình hóa chủ đề (topic modeling) và biểu diễn văn bản dưới dạng vector (word embedding). \
Gensim (viết tắt của "Generate Similar") là một thư viện mã nguồn mở chuyên dùng cho việc xử lý văn bản, đặc biệt là:
- Biến văn bản thành vector số
- Xử lý các tập văn bản lớn
- Mô hình hóa chủ đề (Topic Modeling)
- Tính toán mức độ tương đồng (semantic similarity)

| Tính năng | Mô tả |
| --- | --- |
| Word2Vec |   Huấn luyện mô hình word embedding từ văn bản     |
| Doc2Vec |   Biểu diễn toàn bộ đoạn văn/bài viết thành vector     |
| TF-IDF |   Trích xuất đặc trưng văn bản     |
| LDA, LSI, HDPA |   Mô hình hóa chủ đề (Topic Modeling)     |
| Similarity |   Tính độ tương đồng giữa văn bản     |
| Text Preprocessing |   Tiền xử lý văn bản     |

Một số ứng dụng của Gensim
- Tìm kiếm văn bản theo ngữ nghĩa
- Phân cụm văn bản (clustering)
- Phân loại chủ đề văn bản
- Xây dựng chatbot, hệ thống đề xuất

```python
from gensim import corpora
import nltk
from nltk.tokenize import word_tokenize

# Create sentences list
sentences = [
    "Tôi thích AI",
    "Tôi thích âm nhạc"
]

# Devide sentences into words
nltk.download('punkt_tab')
words = [word_tokenize(sentence) for sentence in sentences]

# Create a dictionary
dictionary = corpora.Dictionary(words)

# Transform sentence into vector bag-of-words
bow_corpus = [dictionary.doc2bow(sentence) for sentence in words]

print(f"Corpus : {bow_corpus}")

# Print vector bag-of-words
for index, vector in enumerate(bow_corpus, 1):
  print(f"Sentence {index} - Vector: {vector}")
```

Word Similarity

```python
from gensim.models import Word2Vec

model = Word2Vec(words, min_count=1)

# Print words similar to "AI"
similar_words = model.wv.most_similar("AI")
print(f'Từ tương tự với "AI": {similar_words}')
```
