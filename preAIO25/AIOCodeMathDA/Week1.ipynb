# Day 1: If/Else

## Bài 1: Tính lịch Can Chi - làm quen với câu lệnh If/Else

• Can Chi là một hệ thống tính toán giờ, ngày, tháng, năm âm lịch của người trung quốc cổ đại. Can Chi có 10 thiên can và 12 địa chi. Tên gọi 10 thiên can Canh, Tân, Nhâm, Quý, Giáp, Ất, Bính, Đinh, Mậu, Kỷ. Tên gọi 12 địa chi Thân, Dậu, Tuất, Hợi, Tý, Sửu, Dần, Mão, Thìn, Tỵ, Ngọ, Mùi.

image.png

```python
def get_can(year):
  num = year % 10
  can = ['Canh', 'Tân', 'Nhâm', 'Quý', 'Giáp', 'Ất', 'Bính', 'Đinh', 'Mậu', 'Kỷ']
  return can[num]

def get_chi(year):
  num = year % 12
  chi = ''
  if num == 0:
    chi = 'Thân'
  elif num == 1:
    chi = 'Dậu'
  elif num == 2:
    chi = 'Tuất'
  elif num == 3:
    chi = 'Hợi'
  elif num == 4:
    chi = 'Tý'
  elif num == 5:
    chi = 'Sửu'
  elif num == 6:
    chi = 'Dần'
  elif num == 7:
    chi = 'Mẹo'
  elif num == 8:
    chi = 'Thìn'
  elif num == 9:
    chi = 'Tỵ'
  elif num == 10:
    chi = 'Ngọ'
  else: # num = 11
    chi = 'Mùi'
  return chi

def calculate_can_chi_calendar(year):
  return get_can(year) + ' ' + get_chi(year)
```

```python
print(f"2025 là năm {calculate_can_chi_calendar(2025)}")
print(f"2024 là năm {calculate_can_chi_calendar(2024)}")
print(f"1997 là năm {calculate_can_chi_calendar(1997)}")
```

## Bài 2: Hàm kích hoạt - Activation function

image.png

1. ReLU (Rectified Linear Unit)
$$ReLU(x)=max(0,x)=\begin{cases}
      x & \text{if $x > 0$}\\
      0 & \text{if $x ≤ 0$}\\ \end{cases}
$$

```python
def relu(x):
  if x > 0:
    return x
  else:
    return 0
```

```python
print(f"relu({3})={relu(3)}")
print(f"relu({-2})={relu(-2)}")
print(f"relu({0})={relu(0)}")
```

2. Leaky ReLU
$$ReLU(x)=max(\alpha x,x)=\begin{cases}
      x & \text{if $x > 0$}\\
      αx & \text{if $x ≤ 0$}\\ \end{cases}
$$

```python
def leaky_relu(x, alpha=0.01):
  if x > 0:
    return x
  else:
    return alpha*x
```

```python
print(f"leaky_relu({3})={leaky_relu(3)}")
print(f"leaky_relu({-2})={leaky_relu(-2)}")
print(f"leaky_relu({0})={leaky_relu(0)}")
```

3. Sigmoid
$$Sigmoid(x)=\dfrac{1}{1+e^{-x}}
$$

```python
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
print(f"sigmoid({0})={sigmoid(0)}")
print(f"sigmoid({-2})={sigmoid(2):.4f}")
print(f"sigmoid({2})={sigmoid(-2):.4f}")
```

4. Tanh
$$
Tanh(x)=\dfrac{e^x-e^{-x}}{e^x+e^{-x}}
$$

```python
import numpy as np
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
print(f"tanh({0})={tanh(0)}")
print(f"tanh({2})={tanh(2):.4f}")
print(f"tanh({150})={tanh(150):.4f}")
```

5. ELU
$$ELU(x)=max( α(e^x-1),x)=\begin{cases}
      x & \text{if $x > 0$}\\
      α(e^x-1) & \text{if $x ≤ 0$}\\ \end{cases}
$$

```python
def elu(x, alpha=1.0):
  if x > 0:
    return x
  else:
    return alpha*(np.exp(x) -1)
```

```python
print(f"elu({3})={elu(3)}")
print(f"elu({-1})={elu(-1):.4f}")
print(f"elu({0})={elu(0):.4f}")
```

+ Số e (còn gọi là hằng số Euler) là một số vô tỷ xuất hiện trong nhiều lĩnh vực toán
học và khoa học, bao gồm cả lĩnh vực tài chính. Trong ngân hàng, số e được sử dụng

tiền gốc để tính lãi cho các kỳ hạn tiếp theo. Công thức tổng quát của số e như sau:
$$
\lim_{x \to \infty } (1+\dfrac{1}{n})^n
$$

+ Giả sử bạn có 1 Dollar, và bạn gởi vào ngân hàng và được nhận lãi mỗi ngày, vậy
điều sẽ xảy ra nếu bạn gởi vào ngân hàng trong 1 năm thì tiền bạn nhận được là bao
nhiêu ? Để trả lời câu hỏi này, chúng ta làm như sau:

- Bước 1: Tiền nhận được sau 1 ngày gửi: $1+\dfrac{1}{365}$
- Bước 2: lặp lại cách tính này cho 1 năm, ta sẽ được công thức của e: $(1+\dfrac{1}{n})^n$

```python
def compute_interest(money, period):
  return (money + money/period)**period
```

## Bài 2: Thuật toán sàng số nguyên tố Eratosthenes

Mô tả: Sử dụng thuật toán sàng Eratosthenes để kiểm tra xem một số nguyên n (được nhập
từ người dùng) có phải là số nguyên tố hay không.

image.png

```python
#%%time
import numpy as np

def is_prime_eratosthenes(n):
  if n < 2:
    return False

  prime = [True] * (n+1)
  prime[0]=prime[1]=False
  #prime[2]=prime[3]=True
  sqrt_int_n = int(np.sqrt(n)) + 1
  for i in range(2, sqrt_int_n,1):
    if prime[i] == True:
      for j in range(i**2, n + 1, i):
        prime[j] = False
  return prime[n]

print(f"is_prime_eratosthenes(10)={is_prime_eratosthenes(10)}")
```

```python
print(f"is_prime_eratosthenes(2)={is_prime_eratosthenes(2)}")
print(f"is_prime_eratosthenes(1)={is_prime_eratosthenes(1)}")
```

# Day 3:  While-Loop

## Bài 1: về sự ngẫu nhiên - Dùng vòng lặp While

Chọn ngẫu nhiên hai số a và b thuộc từ 1 đến 20 sao cho tổng của a + b bằng 40, câu hỏi đặt ra rằng, chúng ta phải tạo ngẫu nhiên bao nhiêu lần để thỏa mãn điều kiện trên ?

```python
import random

start_value = 1
end_value = 20

def random_number_with_condition(total):
  # Set a random value that is the same between devices
  random.seed (0)
  count = 0
  sum = 0
  while(sum != total):
     a = random.randint(start_value, end_value)
     b = random.randint(start_value, end_value)
     sum = a + b
     count += 1
  return count
```

```python
print(f"random_number_with_condition(40)={random_number_with_condition(40)}")
print(f"random_number_with_condition(20)={random_number_with_condition(20)}")
print(f"random_number_with_condition(35)={random_number_with_condition(35)}")
```

• Phương pháp Newton (Newton’s Method), còn được gọi là phương pháp NewtonRaphson, là một phương pháp số học để tìm gần đúng của các nghiệm của một hàm số thực. Cụ thể, nó thường được sử dụng để tìm gần đúng của các nghiệm của phương trình f(x) = 0. \
• Ngoài ứng dụng trong tìm nghiệm của một hàm số, phương pháp Newton còn có ứng dụng trong máy học (Machine learning) trong việc tìm nghiệm của đạo hàm của hàm loss. Tuy nhiên đây là phương pháp không phổ biến bằng thuật toán gradient descent.
